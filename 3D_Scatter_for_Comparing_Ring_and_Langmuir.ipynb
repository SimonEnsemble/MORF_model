Now that we have a solution for dβKR_dβ - dβKL_dβ, we can simplify to find where the Ring MOF outperforms the langmuir model. Our first simplificaiton will be to assume that β=1 in all cases, as no at no point does $\beta$ appear "alone".

```julia
function Leeroy(β,ϵ_Δ,ϵ_♡,δ)
    ϵ = -1/β*log((exp(-β*ϵ_Δ)+exp(-β*(ϵ_♡+δ)))/(1+exp(-β*δ)))
    x = (β*δ*(exp(β*ϵ_Δ) + exp(β*(δ + ϵ_♡)))*exp(β*(2*δ + ϵ + ϵ_Δ + ϵ_♡)) + (β*ϵ - 1)*(exp(β*δ) + 1)^2*exp(2*β*(δ + ϵ_Δ + ϵ_♡)) + (exp(β*δ) + 1)*(-β*(ϵ_Δ*exp(β*(δ + ϵ_♡)) + (δ + ϵ_♡)*exp(β*ϵ_Δ)) + exp(β*ϵ_Δ) + exp(β*(δ + ϵ_♡)))*exp(β*(2*δ + ϵ + ϵ_Δ + ϵ_♡)))*exp(-β*(2*δ + ϵ + 2*ϵ_Δ + 2*ϵ_♡))/(exp(β*δ) + 1)^2
    return x,ϵ
end
res = 100
mag = 5
βϵ_Δ_range = range(0,stop=-1*mag,step=-1*mag/res)
βϵ_♡_range = range(0,stop=-1*mag,step=-1*mag/res)
βδ_range = range(0,stop=mag,step=mag/res)
# worked = zeros(res,res,res)
vals = zeros(res,res,res)

δhold = zeros(0)
Δhold = zeros(0)
♡hold = zeros(0)

for t_index in range(1,stop=res, step=1)
    for h_index in range(1,stop=res, step=1)
        for d_index in range(1,stop=res, step=1)
            if isapprox(Leeroy(1,βϵ_Δ_range[t_index],βϵ_♡_range[h_index],βδ_range[d_index])[1],exp(-1*Leeroy(1,βϵ_Δ_range[t_index],βϵ_♡_range[h_index],βδ_range[d_index])[2]),atol=1)
#                 worked[t_index,h_index,d_index] = 1
                push!(δhold,βδ_range[d_index])
                push!(Δhold,βϵ_Δ_range[t_index])
                push!(♡hold,βϵ_♡_range[h_index])
            end
            vals[t_index,h_index,d_index] = Leeroy(1,βϵ_Δ_range[t_index],βϵ_♡_range[h_index],βδ_range[d_index])[1]-exp(-1*Leeroy(1,βϵ_Δ_range[t_index],βϵ_♡_range[h_index],βδ_range[d_index])[2])
        end
    end
end
# println(δhold)
# println(Δhold)
# println(♡hold)
println(vals)
```
